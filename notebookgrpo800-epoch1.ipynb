```python
import os
import json
import re
import torch
import numpy as np
from tqdm import tqdm
from datasets import Dataset, load_dataset
from sklearn.metrics import classification_report
from transformers import TrainingArguments, AutoTokenizer
from trl import GRPOTrainer
from unsloth import FastLanguageModel
from trl import GRPOConfig 
```

```python
# åŠ è½½é¢„è®­ç»ƒæ¨¡å‹
max_seq_length = 2048
dtype = torch.float16
load_in_4bit = True

print("æ­£åœ¨åŠ è½½æ¨¡å‹...")
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Qwen2.5-7B-Instruct",
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
)
```

```python
# 2. æ•°æ®å‡†å¤‡
# ======================
# åŠ è½½æ•°æ®é›†
HF_TOKEN = "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"
enhanced_dataset = load_dataset(
    "pianjian/pianjian-with-cot-hou",
    data_dir="data",
    split="train[:800]",
    token=HF_TOKEN
)
print(f"å·²åŠ è½½{len(enhanced_dataset)}ä¸ªå¸¦æœ‰æ€ç»´é“¾æ¨ç†çš„ç¤ºä¾‹")
```

```python
# æ ‡ç­¾è½¬æ¢å‡½æ•°
def convert_label(bias_labels):
    return 0 if sum(bias_labels) > 0 else 1

# æç¤ºæ¨¡æ¿
train_prompt_style = """ä¸‹é¢æ˜¯ä¸€ä¸ªæè¿°ä»»åŠ¡çš„è¯´æ˜ï¼Œæä¾›æ›´å¤šä¸Šä¸‹æ–‡çš„è¾“å…¥ã€‚ç¼–å†™é€‚å½“å®Œæˆè¯·æ±‚çš„å“åº”ã€‚

### è¯´æ˜ï¼š
æ‚¨æ˜¯ç¤¾ä¼šåè§åˆ†æä¸“å®¶ï¼Œåœ¨è¯†åˆ«æ–¹é¢å…·æœ‰é«˜çº§çŸ¥è¯†
ä»¥åŠè§£é‡Šä¸æ€§åˆ«ç›¸å…³çš„åˆ»æ¿å°è±¡ã€‚è¯·å®Œæˆä»¥ä¸‹æ­¥éª¤ï¼š
1. åˆ†ææ–‡æœ¬ä¸­æ½œåœ¨çš„æ€§åˆ«åè§
2. ç”Ÿæˆæ€è·¯é“¾è§£é‡Š
3. æä¾›æ‚¨çš„æœ€ç»ˆåå·®åˆ†ç±»ï¼ˆ0 = æœ‰åå·®ï¼Œ1 = ä¸­æ€§ï¼‰

### æ–‡æœ¬ï¼š
{}

### å›åº” ï¼ˆcotï¼‰ï¼š
{}

### Final Answer:
{}"""

# æ ¼å¼åŒ–æ•°æ®é›†
def formatting_prompts_func(examples):
    inputs = examples["original_text"]
    cots = examples["Bias_Analysis_CoT"]
    labels = [convert_label(l) for l in examples["bias_labels"]]
    
    texts = []
    for input_text, cot, label in zip(inputs, cots, labels):
        formatted_text = train_prompt_style.format(
            input_text, cot, label
        ) + tokenizer.eos_token
        texts.append(formatted_text)
    return {"text": texts}

formatted_dataset = enhanced_dataset.map(formatting_prompts_func, batched=True)
```

```python
from peft import LoraConfig, get_peft_model
peft_config = LoraConfig(
    r=32,  # LoRAç§©
    lora_alpha=64,  # ç¼©æ”¾å› å­
    target_modules=["q_proj", "v_proj"],  # ç›®æ ‡æ¨¡å—
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)
model = get_peft_model(model, peft_config)
```

```python
prompt = """ä¸‹é¢æ˜¯ä¸€ä¸ªæè¿°ä»»åŠ¡çš„è¯´æ˜ï¼Œä¸
æä¾›æ›´å¤šä¸Šä¸‹æ–‡çš„è¾“å…¥ã€‚
ç¼–å†™é€‚å½“å®Œæˆè¯·æ±‚çš„å“åº”ã€‚

### è¯´æ˜ï¼š
æ‚¨æ˜¯ç¤¾ä¼šåè§åˆ†æä¸“å®¶ï¼Œåœ¨è¯†åˆ«æ–¹é¢å…·æœ‰é«˜çº§çŸ¥è¯†
ä»¥åŠè§£é‡Šä¸æ€§åˆ«ç›¸å…³çš„åˆ»æ¿å°è±¡ã€‚è¯·å®Œæˆä»¥ä¸‹æ­¥éª¤ï¼š
1. åˆ†ææ–‡æœ¬ä¸­æ½œåœ¨çš„æ€§åˆ«åè§
2. ç”Ÿæˆæ€è·¯é“¾è§£é‡Š
3. æä¾›æ‚¨çš„æœ€ç»ˆåå·®åˆ†ç±»ï¼ˆ0 = æœ‰åå·®ï¼Œ1 = ä¸­æ€§ï¼‰

### æ–‡æœ¬ï¼š
{}

### å›åº” ï¼ˆæ€è·¯é“¾ï¼‰ï¼š
{}

"""
```

```python
# ======================
# 3. GRPOå¼ºåŒ–å­¦ä¹ å®ç°
# ======================
def parse_predicted_labels(response):
    """è§£ææ¨¡å‹è¾“å‡ºçš„ä¸‰ç»´æ ‡ç­¾"""
    # å°è¯•åŒ¹é…[0,1,0]æ ¼å¼
    bracket_match = re.search(r'$$([01]),\s*([01]),\s*([01])$$', response)
    if bracket_match:
        return [int(x) for x in bracket_match.groups()]
    
    # å¤‡é€‰æ–¹æ¡ˆï¼šæå–æ‰€æœ‰0/1æ•°å­—
    digits = re.findall(r'\b[01]\b', response)
    if len(digits) >= 3:
        return [int(x) for x in digits[:3]]
    
    return [0, 0, 0]  # é»˜è®¤å€¼


# ä¿®æ”¹è®­ç»ƒå‚æ•°é…ç½®
# æ­£ç¡®é…ç½® GRPO å‚æ•°
training_args = GRPOConfig(
    output_dir="./grpo_results",
    per_device_train_batch_size=2,  # ä¿æŒå°æ‰¹é‡
    gradient_accumulation_steps=1,  # å¢å¤§ç´¯ç§¯æ­¥æ•°
    
    # å…³é”®ä¿®å¤å‚æ•°
    num_generations=2,              # å¤§å¹…å‡å°‘ç”Ÿæˆæ•°é‡
    max_completion_length=80,       # ç¼©çŸ­ç”Ÿæˆé•¿åº¦
    use_vllm=False,                 # ç¦ç”¨vLLM
    bf16=False,                     # ç¦ç”¨BF16
    fp16=True,                      # å¯ç”¨FP16
    
    # ç¨³å®šæ€§å¢å¼º
    gradient_checkpointing=True,     # å¯ç”¨æ¢¯åº¦æ£€æŸ¥ç‚¹
    max_grad_norm=0.3,              # å‡å°æ¢¯åº¦è£å‰ª
    
    # è®­ç»ƒæ§åˆ¶
    learning_rate=1e-5,
    num_train_epochs=1,
    logging_steps=5,
    save_steps=100,
    eval_strategy="no",
    # eval_steps=50,
)

```

```python
# 1. é‡å†™æ•°æ®æ ¼å¼åŒ–å‡½æ•°
def formatting_prompts_func(examples):
    inputs = examples["original_text"]
    cots = examples["Bias_Analysis_CoT"]
    labels = [convert_label(l) for l in examples["bias_labels"]]
    
    prompts = []
    responses = []
    for input_text, cot, label in zip(inputs, cots, labels):
        # åˆ†ç¦»promptï¼ˆæŒ‡ä»¤+æ–‡æœ¬ï¼‰å’Œresponseï¼ˆæ¨ç†é“¾+ç­”æ¡ˆï¼‰
        prompt_text = train_prompt_style.format(input_text, "", "").split("### å›åº” ï¼ˆcotï¼‰ï¼š")[0].strip()
        response_text = f"{cot}\n\n### Final Answer:\n{label}"
        
        prompts.append(prompt_text)
        responses.append(response_text)
    
    return {"prompt": prompts, "response": responses}  # å…³é”®ï¼šè¿”å›åŒå­—æ®µå­—å…¸

# 2. ä¿®æ”¹å¥–åŠ±å‡½æ•°é€‚é…æ–°æ ¼å¼
def reward_function(completions, prompts, **kwargs):
    """ç¬¦åˆGRPOè§„èŒƒçš„å¥–åŠ±å‡½æ•°è®¾è®¡"""
    rewards = []
    for i in range(len(completions)):
        completion = completions[i]
        prompt = prompts[i]
        
        # ä»promptä¸­æå–çœŸå®æ ‡ç­¾
        true_label_match = re.search(r'### çœŸå®æ ‡ç­¾:\s*($$[0-1],[0-1],[0-1]$$)', prompt)
        if true_label_match:
            try:
                true_labels = json.loads(true_label_match.group(1))
            except:
                true_labels = [0, 0, 0]
        else:
            true_labels = [0, 0, 0]
        
        # è§£æé¢„æµ‹æ ‡ç­¾
        pred_labels = parse_predicted_labels(completion)
        
        # è®¡ç®—åŸºç¡€å¥–åŠ±ï¼šåˆ†ç±»æ­£ç¡®+1ï¼Œé”™è¯¯-1
        base_reward = 1.0 if pred_labels == true_labels else -1.0
        
        # æ ¼å¼è§„èŒƒå¥–åŠ±ï¼šè¾“å‡ºåŒ…å«æ–¹æ‹¬å·æ ¼å¼+0.2
        format_reward = 0.2 if "[" in completion and "]" in completion else 0.0
        
        # æ€ç»´é“¾å¥–åŠ±ï¼šåŒ…å«æ¨ç†æ­¥éª¤+0.3
        cot_reward = 0.3 if "æ€è·¯é“¾" in completion or "åˆ†ææ­¥éª¤" in completion else 0.0
        
        total_reward = base_reward + format_reward + cot_reward
        rewards.append(total_reward)
    
    return rewards

# 3. åº”ç”¨æ–°æ ¼å¼åŒ–å‡½æ•°
formatted_dataset = enhanced_dataset.map(formatting_prompts_func, batched=True)

# 4. éªŒè¯æ•°æ®é›†æ ¼å¼ï¼ˆå¿…éœ€ï¼ï¼‰
assert "prompt" in formatted_dataset[0], "æ•°æ®é›†ç¼ºå°‘promptå­—æ®µ"
assert "response" in formatted_dataset[0], "æ•°æ®é›†ç¼ºå°‘responseå­—æ®µ"
print("âœ… æ•°æ®é›†æ ¼å¼éªŒè¯é€šè¿‡")
grpo_trainer = GRPOTrainer(
    model=model,                    # ç­–ç•¥æ¨¡å‹
    processing_class=tokenizer,     # å…³é”®ä¿®æ­£ï¼šä½¿ç”¨ processing_class ä»£æ›¿ tokenizer [6,9](@ref)
    reward_funcs=reward_function,   # å¥–åŠ±å‡½æ•° [1](@ref)
    args=training_args,             # è®­ç»ƒé…ç½®
    train_dataset=formatted_dataset, # è®­ç»ƒæ•°æ®
)
```

```python
def load_data(biased_path, neutral_path):
    """åŠ è½½å¹¶è‡ªåŠ¨å¹³è¡¡æ•°æ®é›†"""
    from datasets import Dataset, concatenate_datasets
    import json

    # åŠ è½½åè§æ•°æ®
    with open(biased_path, 'r', encoding='utf-8-sig') as f:
        biased_data = [
            {
                'original_text': item['ori_sentence'],
                'edited_text': item['edit_sentence'],
                'bias_labels': item['bias_labels'],
                'text_type': 'biased'
            }
            for item in json.load(f)
        ]
    
    # åŠ è½½ä¸­æ€§æ•°æ®
    with open(neutral_path, 'r', encoding='utf-8-sig') as f:
        neutral_data = [
            {
                'original_text': item['text'],
                'edited_text': None,
                'bias_labels': [0, 0, 0],
                'text_type': 'neutral'
            }
            for item in json.load(f)
        ]

    # åˆ›å»ºåˆ†å±‚æ•°æ®é›†
    biased_dataset = Dataset.from_list(biased_data)
    neutral_dataset = Dataset.from_list(neutral_data)
    
    # è‡ªåŠ¨å¹³è¡¡é‡‡æ ·
    min_size = min(len(biased_dataset), len(neutral_dataset))
    balanced_dataset = concatenate_datasets([
        biased_dataset.shuffle().select(range(min_size)),
        neutral_dataset.shuffle().select(range(min_size))
    ]).shuffle()
    
    return balanced_dataset
```

```python
import json
import re
import numpy as np
from tqdm import tqdm
from sklearn.metrics import classification_report

def evaluate_val_performance(model, tokenizer, balanced_dataset, output_dir=None,test_mode=True):
    """ç»¼åˆè¯„ä¼°ä»»åŠ¡1ï¼ˆæ•´ä½“åè§ï¼‰å’Œä»»åŠ¡2ï¼ˆå­ç±»åˆ«åˆ†ç±»ï¼‰"""
    # æ•°æ®å‡†å¤‡
    dataset = balanced_dataset.to_pandas().to_dict('records')
    if test_mode:
        dataset = dataset[:2]  # ä»…å–å‰5ä¸ªæ ·æœ¬
    # åˆå§‹åŒ–å­˜å‚¨
    task1_true, task1_pred = [], []  # ä»»åŠ¡1æ ‡ç­¾
    task2_true, task2_pred = [], []  # ä»»åŠ¡2æ ‡ç­¾ï¼ˆä¸‰ç»´ï¼‰
    results = []

    # æç¤ºæ¨¡æ¿ï¼ˆä¸è®­ç»ƒä¸€è‡´ï¼‰
    classification_template = """ä»¥ä¸‹æ˜¯æè¿°ä»»åŠ¡çš„è¯´æ˜ã€‚å†™ä¸€ä¸ªé€‚å½“åœ°å®Œæˆè¯·æ±‚çš„å›å¤ã€‚
    ### è¯´æ˜ï¼šä½œä¸ºæ€§åˆ«åè§åˆ†æå¸ˆï¼Œè¯·åˆ†æä»¥ä¸‹æ‰€æœ‰ä¸‰ç§åè§ç±»å‹çš„æ–‡æœ¬ã€‚
    æ¯ç§ç±»å‹éƒ½å¿…é¡»ç”¨ 0ï¼ˆå¦ï¼‰æˆ– 1ï¼ˆæ˜¯ï¼‰æ¥å›ç­”ï¼š
    1. åˆ»æ¿å°è±¡ï¼ˆä¾‹å¦‚ï¼Œæ ¹æ®æ€§åˆ«å‡è®¾ç‰¹å¾ï¼‰
    2. èƒ½åŠ›å‡è®¾ï¼ˆä¾‹å¦‚ï¼Œä¸èƒ½åŠ›ç›¸å…³çš„åˆ»æ¿å°è±¡ï¼‰
    3. ç¤¾ä¼šè§’è‰²åè§ï¼ˆä¾‹å¦‚ï¼ŒèŒä¸š/è§’è‰²æœŸæœ›ï¼‰
    [é‡è¦]ä½ çš„æœ€ç»ˆç­”æ¡ˆå¿…é¡»æ˜¯ç”¨æ–¹æ‹¬å·æ‹¬èµ·æ¥çš„ä¸‰ä¸ª 0/1 æ•°å­—çš„åˆ—è¡¨ã€‚
    ### æ–‡æœ¬ï¼š{input_text}
    ### åˆ†ææ­¥éª¤ï¼šaï¼‰ è¯†åˆ«ä¸æ€§åˆ«ç›¸å…³çš„æœ¯è¯­bï¼‰ æ£€æŸ¥æ¯ç§åè§ç±»å‹cï¼‰ åˆ¶å®šæœ€ç»ˆç»“è®º

### Final Answer:
"""

    # å¤„ç†æ¯ä¸ªæ ·æœ¬
    for item in tqdm(dataset, desc="Processing Samples"):
        try:
            input_text = item['original_text']
            
            # ================= æ ‡ç­¾å¤„ç† =================
            # ä»»åŠ¡1æ ‡ç­¾ï¼ˆæ•´ä½“åè§ï¼‰
            task1_true.append(1 if any(item['bias_labels']) else 0)
            
            # ä»»åŠ¡2æ ‡ç­¾ï¼ˆå­ç±»åˆ«åˆ†ç±»ï¼‰
            task2_true.append(item['bias_labels'])  # æ ¼å¼ï¼š[AC, DI, ANB]
            
            # ================= æ¨¡å‹é¢„æµ‹ =================
            prompt = classification_template.format(input_text=input_text)
            inputs = tokenizer(prompt, return_tensors="pt", 
                              max_length=1024, truncation=True).to("cuda")
            outputs = model.generate(
                input_ids=inputs.input_ids,
                max_new_tokens=200,
                temperature=0.1,
                do_sample=False
            )
            response = tokenizer.decode(outputs[0], skip_special_tokens=True)
            
            # --- ç²¾å‡†è§£æé€»è¾‘ ---
            labels = [0, 0, 0]
            
            # å®šä½Final Answeréƒ¨åˆ†
            final_section = ""
            if "### Final Answer:" in response:
                final_section = response.split("### Final Answer:")[-1].strip()
            elif "Final Answer" in response:
                final_section = response.split("Final Answer")[-1].strip()
            
            # ä¼˜å…ˆè§£ææ–¹æ‹¬å·æ ¼å¼
            bracket_match = re.search(r'$$\s*([01])\s*,\s*([01])\s*,\s*([01])\s*$$', final_section)
            if bracket_match:
                labels = list(map(int, bracket_match.groups()))
            else:
                # å¤‡é€‰æ–¹æ¡ˆï¼šæå–æ‰€æœ‰0/1æ•°å­—
                digits = re.findall(r'\b[01]\b', final_section)
                if len(digits) >= 3:
                    pred_labels = list(map(int, digits[:3]))
            pred_labels = pred_labels[:3] + [0]*(3-len(pred_labels))

            
            # å­˜å‚¨é¢„æµ‹ç»“æœ
            task1_pred.append(1 if any(pred_labels) else 0)
            task2_pred.append(pred_labels)
            
            # è®°å½•è¯¦ç»†ç»“æœ
            results.append({
                "text": input_text,
                "true_labels": item['bias_labels'],
                "pred_labels": pred_labels,
                "response": response
            })
            
        except Exception as e:
            print(f"å¤„ç†å¤±è´¥: {str(e)}")
            task1_pred.append(0)
            task2_pred.append([0,0,0])

    # ================= æŒ‡æ ‡è®¡ç®— =================
    print("\n" + "="*60 + "\nè¯¦ç»†æ ·æœ¬çº§é¢„æµ‹ç»“æœ:")
    for i, res in enumerate(results):
        print(f"\næ ·æœ¬ {i+1}:")
        print(f"åŸæ–‡: {res['text']}")
        print(f"çœŸå®æ ‡ç­¾: {res['true_labels']}")
        print(f"é¢„æµ‹æ ‡ç­¾: {res['pred_labels']}")
        print(f"æ¨¡å‹å“åº”:\n{res['response']}\n" + "-"*50)
    print("="*60 + "\n")
    # ä»»åŠ¡1ï¼šäºŒåˆ†ç±»æŒ‡æ ‡
    task1_report = classification_report(
        task1_true, task1_pred,
        target_names=['Neutral', 'Biased'],
        output_dict=True
    )
    
    # ä»»åŠ¡2ï¼šå¤šæ ‡ç­¾åˆ†ç±»ï¼ˆå®å¹³å‡ï¼‰
    task2_report = classification_report(
        np.array(task2_true),
        np.array(task2_pred),
        target_names=['AC', 'DI', 'ANB'],
        output_dict=True,
        zero_division=0
    )
    
    # ================= ç»“æœè¾“å‡º =================
    print("\n" + "="*60)
    print("ä»»åŠ¡1 - æ•´ä½“åè§æ£€æµ‹:")
    print(f"Precision: {task1_report['Biased']['precision']:.4f}")
    print(f"Recall:    {task1_report['Biased']['recall']:.4f}")
    print(f"F1-score:  {task1_report['Biased']['f1-score']:.4f}")
    
    print("\n" + "="*60)
    print("ä»»åŠ¡2 - å­ç±»åˆ«åˆ†ç±»ï¼ˆå®å¹³å‡ï¼‰:")
    print(f"Precision: {task2_report['macro avg']['precision']:.4f}")
    print(f"Recall:    {task2_report['macro avg']['recall']:.4f}")
    print(f"F1-score:  {task2_report['macro avg']['f1-score']:.4f}")
    
    print("\nå„å­ç±»åˆ«è¯¦ç»†æŒ‡æ ‡:")
    for cls in ['AC', 'DI', 'ANB']:
        print(f"{cls}: P={task2_report[cls]['precision']:.4f} R={task2_report[cls]['recall']:.4f} F1={task2_report[cls]['f1-score']:.4f}")
    print("="*60 + "\n")



# ä½¿ç”¨ç¤ºä¾‹ --------------------------------------------------
balanced_val = load_data(
    biased_path="/kaggle/input/pianjian/yanzheng biased.json",
    neutral_path="/kaggle/input/pianjian/yanzhneg non-biased.json"
)

# è¿è¡Œè¯„ä¼°
report = evaluate_val_performance(
    model=model,
    tokenizer=tokenizer,
    balanced_dataset=balanced_val,
    output_dir="./eval_results"
)
```
